```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("students_data.csv")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.columns = df.columns.str.lower().str.strip()
df
```

```python
df.replace(["missing", "NaN"], np.nan, inplace=True)
df
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["age"] = df["age"].fillna(df["age"].median()).astype(int)
```

```python
df["math_score"] = pd.to_numeric(df["math_score"], errors="coerce")
df["english_score"] = pd.to_numeric(df["english_score"], errors="coerce")

df["math_score"] = df["math_score"].fillna(df["math_score"].mean()).astype(int)
df["english_score"] = df["english_score"].fillna(df["english_score"].mean()).astype(int)
```

```python
df['gender'] = df['gender'].str.lower().map({
    'male': 'M',
    'female': 'F'

})
df
```

```python
df['grade'] = df['grade'].astype(str)

df['grade'] = df['grade'].replace({
    '10th': 'X',
    '11th': 'XI',
    '12': 'XII',
    '11': 'XI',
    '10': 'X'
})
df
```

```python
# Convert ENROLLED_DATE to datetime (fix mixed formats)
df["enrolled_date"] = pd.to_datetime(
    df["enrolled_date"],
    errors="coerce",
    dayfirst=True
)

# Change format to DD-MM-YYYY
df["enrolled_date"] = df["enrolled_date"].dt.strftime("%d-%m-%Y")
```

```python
df["enrolled_date"] = df["enrolled_date"].fillna("01-01-2022")
```

```python
df['name'] = df['name'].str.strip()
df['gender'] = df['gender'].str.strip()
df['remarks'] = df['remarks'].str.strip()
```

```python
df['name'] = df['name'].str.upper()
df['gender'] = df['gender'].str.upper()
df['remarks'] = df['remarks'].str.upper()
```

```python
# Rename Columns
df.columns = df.columns.str.strip().str.upper()

```
